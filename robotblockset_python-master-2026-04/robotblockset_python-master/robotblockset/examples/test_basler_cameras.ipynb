```python
import cv2
import mediapy as media
from robotblockset.cameras.basler import BaslerRGBCamera, list_basler_cameras
```

```python
cam = BaslerRGBCamera(serial_number="22084405")
cam.connect()
```

```python
while True:
    img_u8 = cam.get_rgb_image_as_int()
    cv2.imshow("basler", cv2.cvtColor(img_u8, cv2.COLOR_RGB2BGR))
    if cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()
```
