```python
from robotblockset.tools import get_rbs_path, print_xml, print_xml_for_console

import mujoco 
from robotblockset.mujoco.tools_mjcf import print_body_tree, actuators_for_joint
```

```python
PLATFORM = "tiagobase"
ROBOT = "panda"
GRIPPER = None
```

```python
MODEL_PATH = get_rbs_path() + "/mujoco/mjcf_models/"
print(MODEL_PATH)
```

```python
scene=mujoco.MjSpec.from_file(MODEL_PATH + "scene.xml")
scene.copy_during_attach = True
# print_xml(scene.to_xml())
```

```python
platform_spec = mujoco.MjSpec.from_file(MODEL_PATH + f"{PLATFORM}.xml")
platform_spec.copy_during_attach = True
print_body_tree(platform_spec.worldbody, platform_spec)
# print_xml(platform_spec.to_xml())
```

```python
attachment_frame = scene.worldbody.add_frame()
attachment_frame.attach_body(platform_spec.body("base"), f"{PLATFORM}_")
scene.body(f"{PLATFORM}_base").name = PLATFORM
print_body_tree(scene.worldbody, scene)
# print_xml(scene.to_xml())
```

```python
if ROBOT is not None:
    robot_spec = mujoco.MjSpec.from_file(MODEL_PATH + f"{ROBOT}.xml")
    robot_spec.copy_during_attach = True
    print_body_tree(robot_spec.worldbody, robot_spec)
    # print_xml(robot_spec.to_xml())
```

```python
if ROBOT is not None:
    attachment_frame = scene.body(f"{PLATFORM}_mounting_base").add_frame(pos=[0.1, 0, 0])
    attachment_frame.attach_body(robot_spec.body("base"), f'{ROBOT}_')
    scene.body(f"{ROBOT}_base").name = ROBOT
    print_body_tree(scene.worldbody, scene)
    # print_xml(scene.to_xml())
```

```python
if GRIPPER is not None:
    gripper_spec = mujoco.MjSpec.from_file(MODEL_PATH + f"{GRIPPER}.xml")
    gripper_spec.copy_during_attach = True
    attachment_frame = scene.body(f"{ROBOT}_flange").add_frame(pos=[0, 0, 0], quat=[0.9238795, 0, 0, -0.3826834])
    attachment_frame.attach_body(gripper_spec.body("hand"), f'{ROBOT}_tool_')
    scene.delete(scene.site(f"{ROBOT}_TCP"))
    scene.site(f"{ROBOT}_tool_hand_TCP").name = f"{ROBOT}_TCP"
print_body_tree(scene.worldbody, scene)
```

```python
for actuator in scene.actuators:
  print(actuator.name)
```

```python
if GRIPPER is None:
    if ROBOT is not None:
        scene.modelname = f"{PLATFORM}_{ROBOT}_scene"
    else:
        scene.modelname = f"{PLATFORM}_scene"
else:
    scene.modelname = f"{PLATFORM}_{ROBOT}_{GRIPPER}_scene"
scene.option.timestep = 0.001
```

```python
print_body_tree(scene.worldbody, scene)
```

```python
# print_xml(scene.to_xml())
```

```python
if ROBOT is not None:
    with open(MODEL_PATH + f"{PLATFORM}_{ROBOT}_scene.xml", "w") as f:
        f.write(scene.to_xml())
else:
    with open(MODEL_PATH + f"{PLATFORM}_scene.xml", "w") as f:
        f.write(scene.to_xml())
```
